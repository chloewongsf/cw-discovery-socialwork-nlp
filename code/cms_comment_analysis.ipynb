```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

template_text = """
i write to you, as a social worker, to comment on the proposed rule addressing nurse minimum staffing standards for long-term care facilities and medicaid institutional payment transparency reporting (cms–3442–p). social workers and interdisciplinary colleagues—including certified nursing assistants (cnas), licensed practical nurses–licensed vocational nurses (lpns–lvns), and registered nurses (rns)—strive to enhance the quality of life and quality of care for nursing home residents. yet, their efforts are often hindered by exceedingly high workloads. i appreciate cms’s efforts to establish minimum staffing standards for nursing disciplines. cms’s proposed minimum nurse staffing standards of 2.45 hours per resident day (hprd) of nurse aide care and 0.55 hprd of rn care improve upon current regulation, which defers to each facility to determine “sufficient” staffing levels. however, these proposed standards are much lower than those supported by more than 20 years of research, and they would allow for a reduction in total staffing to levels below both prepandemic and may 2022 national staffing averages. i urge cms to strengthen the proposed minimums by (1) increasing the minimum nurse aide standard to 2.45 cna hprd and specifying that the preceding standard applies specifically to cnas, who meet federal training requirements and have state certifications that nurse aides may lack; (2) increasing the minimum rn staffing requirement to 0.75 hprd; (3) adding a minimum standard of 1.4 licensed nursing (total rn and lpn–lvn) hprd, thereby recognizing the unique contribution of each licensed nursing discipline to resident care; and (4) adding a total nursing staff minimum standard of 4.2 hprd. cms’s proposal to require an rn onsite 24 hours per day, seven days per week is also a great improvement on the current regulation. i suggest that cms strengthen the language by (1) requiring 24–7 rns who provide care directly to residents onsite, not on call or as needed, and (2) clarifying that facilities cannot use nurses in administrative roles, such as director of nursing or mds coordinator, to meet 24–7 rn direct care requirement. the implementation and enforcement of a nurse minimum staffing standard is vital both to the biopsychosocial well-being of nursing home residents and to the safety and well-being of staff. i encourage cms to strengthen its implementation plan in two ways: 1. eliminate hardship exemptions/waivers for facilities that cannot meet minimum nurse staffing standards, the 24–7 requirement, or both. every nursing home resident deserves high-quality care, regardless of their geographic location or other factors. 2. implement all nurse staffing requirements within two years of the final rule and on the same timeline, regardless of geographic area; a phase-in of three to seven years is too long. establish reasonable benchmarks for each year to help facilities meet the two-year goal. 3. if a nursing home is unable to meet the minimum nurse staffing standards, place a hold on payment for new admissions until the problem is remedied. i also urge cms to strengthen the federal requirements for social service staff in nursing homes. a 2022 report on nursing homes by the national academies of sciences, engineering, and medicine recommended that every nursing home in the country hire at least one social worker, defined as a bsw or msw and excluding any other academic discipline in the “human services field." staffing ratios for social workers also need to be improved; multiple studies have shown that the “120-bed rule” is insufficient. consideration of the distinct and equally important needs of residents receiving and those in long-term care is essential in determining more appropriate staffing ratios. every resident of every nursing home deserves access to a social worker, as defined by nasw and the national academies, and to sufficient nursing care at all levels. thank you for your leadership and your consideration of my input.
"""
documents = [template_text] + df['Comment'].tolist()  # Template first, then each comment

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(documents)

cosine_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:])[0]  # Only compare template to comments

df['similarity_to_template'] = cosine_sim

template_matches = df[df['similarity_to_template'] > 0.8]

print(template_matches[['Comment', 'similarity_to_template']])
```
